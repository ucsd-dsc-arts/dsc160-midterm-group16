```python
from bs4 import BeautifulSoup
import os
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_songs_recorded_by_the_Beatles"
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text, 'lxml')
```

```python
links = soup.find_all("table")[2]
```

```python
links = soup.find_all("table")[2]
links = links.find_all("a")
for x in range(len(links)):
    links[x] = links[x].get('href')
    
wikis = links
```

```python
rows = soup.find_all("table")[2].find_all("tr")
```

```python
rows[1].find_all('a')[0].get('href')
```

```python
songs = []
for y in range(1,len(rows)):
    wiki = rows[y].find_all('a')[0].get('href')
    songs.append(wiki)
```

```python
wikipedia = "https://en.wikipedia.org"
songs[0]
```

```python
song_page = requests.get(wikipedia + songs[0])
```

```python
spage = BeautifulSoup(song_page.text, 'lxml')
```

```python
spage.find_all('table')[0].find_all('a')
```

```python
genres = []
link = []
for z in range(len(songs)):
    song_page = requests.get(wikipedia + songs[z])
    spage = BeautifulSoup(song_page.text, 'lxml')
    spage = spage.find_all('table')[0].find_all('a')
    for w in range(len(spage)):
        if spage[w].get('href') ==  '/wiki/Music_genre':
            genres.append(spage[w+1].text)
            link.append(songs[z])
```

```python
song = []
for x in range(1,len(rows)):
    if rows[x].find_all('a')[0].get('href') in link:
        song.append(rows[x].find_all('a')[0].get('title'))
len(song)
```

```python
import pandas as pd
```

```python
genre_data = pd.DataFrame({"Song":song,"Genre":genres})
```

```python
genre_data.to_csv('genre.csv',index=False)
```
